畳み込みニューラルネットワークの概念

なおストライドとパディングは1とする。

各辺について

(入力画像のサイズ + パディング \* 2 + ストライド - フィルタサイズ)/(ストライド) = 出力画像サイズ

であるため、

(6 + 1 * 2 + 1 - 2)/ 1 = 7

よって出力画像は7*7
