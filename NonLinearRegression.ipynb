非線形回帰モデル

```python
#numpyと、ワインのデータセット読み込み
import numpy as np
from sklearn.datasets import load_wine
from pandas import DataFrame
#wine変数にデータ読み込み
wine = load_wine()
#wineデータをデータフレームに変換
df = DataFrame(data=wine.data, columns = wine.feature_names)
```

```python
#データフレーム表示
df.head(30)
```

```python
#項目名表示
print(wine['feature_names'])
```

```python
#flavanoids列のみflavに抽出
flav = df.loc[:, 'flavanoids'].values
#prolineのみproに抽出
pro = df.loc[:, 'proline'].values
```

```python
#描画ファイルインポート
import matplotlib.pyplot as plot
```

```python
#点で描画
plot.plot(flav,pro,'o')
```

```python
# 近似式の係数を算出
np.polyfit(flav, pro, 2)
```

```python
#点と非線形近似式の同時描写
plot.plot(flav,pro,'o')
plot.plot(flav,np.poly1d(np.polyfit(flav, pro, 2))(flav))
```
