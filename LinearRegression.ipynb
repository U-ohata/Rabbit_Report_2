線形回帰モデル

```python
#numpyと、ボストンの住宅データセット読み込み
import numpy as np
from sklearn.datasets import load_boston
from pandas import DataFrame
```

```python
#boston変数にデータ読み込み
boston = load_boston()
```

```python
#bostonデータ表示
print(boston)
```

```python
#目的変数表示
print(boston['target'])
```

```python
#説明変数表示
print(boston['data'])
```

```python
#bostonデータをデータフレームに変換
df = DataFrame(data=boston.data, columns = boston.feature_names)
```

```python
#データフレーム表示
df.head(30)
```

```python
#データフレームに価格を追加
df['PRICE']=np.array(boston.target)
```

```python
#データフレーム再度表示(PRICE追加)
df.head(30)
```

```python
#RM列のみdataに抽出
data = df.loc[:, ['RM']].values
```

```python
#RM列のみ表示
data[0:30]
```

```python
#価格のみ抽出、表示
target = df.loc[:, 'PRICE'].values
target[0:5]
```

```python
#線形回帰モデルインポート
from sklearn.linear_model import LinearRegression
```

```python
#model変数に線形回帰モデル代入、
model = LinearRegression()
model.get_params()
```

```python
#モデルに学習用データを充てる
model.fit(data,target)
```

```python
#実際の予測(RM入力、価格出力)
model.predict([[10]])
```

```python
#CRIMとRMのみ抽出し、target2に価格抽出
data2 = df.loc[:, ['CRIM','RM']].values
target2 = df.loc[:, 'PRICE'].values
```

```python
#data2の中身を5つ表示
print(data2[0:5])
```

```python
#model2に線形回帰モデルを充てて、学習
model2 = LinearRegression()
model2.fit(data2,target2)
```

```python
#CRIMとRMを代入して出力
model2.predict([[0.6,7]])
```
